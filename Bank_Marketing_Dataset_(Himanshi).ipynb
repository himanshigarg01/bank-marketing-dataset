```python
# important libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df = pd.read_csv('/content/bank-full.csv', delimiter = ';')
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=35)
x_os, y_os = smote.fit_resample(df.iloc[:,0:16], df.iloc[:,-1])
print(x_os)
```

```python
# split into train and test
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,0:16], df.iloc[:,-1], test_size = 0.2, random_state = 35)
print(x_train.shape, y_train.shape)
```

```python
ohe = OneHotEncoder(drop = 'first')
x_transformed = ohe.fit_transform(x_train[['marital','job', 'education']])
print(x_transformed)
```

```python
ohe = OneHotEncoder(drop = 'first')
x_transformed = ohe.fit_transform(x_train[['marital','job', 'education']])
print(x_transformed)
```

```python
ohe = OneHotEncoder(drop = 'first')
x_test_trans = ohe.fit_transform(x_test[['marital','job', 'education']])
print(x_test_trans)
```

```python
lb = LabelEncoder()
y_train_enc = lb.fit_transform(y_train)
y_test_enc = lb.fit_transform(y_test)
```

```python
model = RandomForestClassifier(random_state=42, )
model.fit(x_transformed, y_train_enc)
y_pred = model.predict(x_test_trans)
```

```python
accuracy = accuracy_score(y_test_enc, y_pred)
print(accuracy)
```

```python
report = classification_report(y_test_enc, y_pred)
print(report)
```
